```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

favor = pd.Series(['WINTER', 'SPRING', 'SUMMER', 'FALL','SUMMER','SPRING','FALL','SPRING','SUMMER'])
fd = favor.value_counts()

fd.plot.barh(xlabel='Frequency', 
            ylabel='Season', 
            rot=45, 
            title='Favorite Season')

plt.subplots_adjust(left=0.2) # 그래프의 왼쪽 여백을 조정
plt.show()
```

```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

favor = pd.Series(['WINTER', 'SPRING', 'SUMMER', 'FALL','SUMMER','SPRING','FALL','SPRING','SUMMER'])
fd = favor.value_counts()

fd.plot.pie(ylabel='', # y라벨을 공백으로 한 이유는? 값을 넣으면 불필요한 문자열 표시되기 때문
            autopct='%1.0f%%', #백분율 출력 지정
            title='Favorite Season',)
plt.show()
```

```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

colors = pd.Series([2,3,2,1,1,2,2,1,3,2,1,3,2,1,2])
fd = colors.value_counts()  

fd.index = ['Red', 'Blue', 'Green'] # 인덱스 이름 변경
fd.index

fd.plot.bar(xlabel='Color', 
            ylabel='Frequency',
            rot=40, 
            title='Favorite Color',)
plt.show()

fd.plot.pie(ylabel='',
            autopct='%1.1f%%', 
            title='Favorite Color',)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cars.csv', encoding='utf-8')
dist = df['dist']


# 구간 개수를 지정해서 히스토그램
dist.plot.hist()
plt.show()

dist.plot.hist(bins=6) # 막대 개수 지정
plt.show()


# 구간별 반도수 계산
dist.value_counts(bins=5, sort=False)
dist.plot.hist(bins=6,
               xlabel='Distance',
                ylabel='Frequency',
                title='Barking distance',
                color='g')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('iris.csv', encoding='utf-8')

df.boxplot(column='Sepal_Length', # 상자 그림 대상 컬럼 지정
           by='Species',         # 그룹정보 컬럼
           grid=False)        # 격자선 제거)

plt.suptitle=('')
plt.show()

df.boxplot(column='Sepal_Width', # 상자 그림 대상 컬럼 지정
           by='Species',         # 그룹정보 컬럼
           grid=True)        # 격자선 제거)

plt.suptitle=('')
plt.show()

df.boxplot(column='Sepal_Width', # 상자 그림 대상 컬럼 지정
           by='Species',         # 그룹정보 컬럼
           grid=False)        # 격자선 제거)

plt.suptitle=('')
plt.show()

df.boxplot(column='Sepal_Width', # 상자 그림 대상 컬럼 지정
           by='Species',         # 그룹정보 컬럼
           grid=False)        # 격자선 제거)

plt.suptitle=('')
plt.show()
```
